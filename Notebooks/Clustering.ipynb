```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import metrics
import random
```

```python
df = pd.read_csv("df_users.csv")
df_genre = df.drop(["Unnamed: 0"], axis=1)
df_genre.head()
```

```python
df_genre = df_genre.groupby("Id_user").sum()
df_genre.head()
```

```python
kmeanModel = KMeans(n_clusters=17)
kmeanModel.fit(df_genre)
```

```python
df_genre["cluster"] = kmeanModel.labels_
# df_genre["cluster"].sample(n=10)
df_genre = df_genre[["cluster"]].copy()
df_genre.value_counts()
```

```python
df_genre.to_csv("clusters_of_users_by_genre.csv")
```
